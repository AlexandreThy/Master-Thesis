```python
def dottau(tau,t,u,test):
    return (u(t) - tau)/0.06



# We want to evaluate the system on 30 linearly
# spaced times between t=0 and t=3.
def u(t):
    if t > 0.2 and t < 0.7:
        return 1
    return 0
t = np.linspace(0, 1, 600)
my_u = np.zeros(600)
for i in range(600):
    my_u[i] = u(t[i])


# We simulate the system for different values of k.
    # We simulate the system and evaluate $v$ on the
    # given times.
v = scipy.integrate.odeint(dottau, [0,0], t,args=(u,0))
print(v.shape)
u = v[:,0]
tau = v[:,1]

    # We plot the particle's trajectory.
fig = go.Figure()
fig.add_trace(go.Scatter(x=t, y=tau.reshape(600),
                    mode='lines',
                    name='Joint Torque'))
fig.add_trace(go.Scatter(x=t, y=my_u,
                    mode='lines',
                    name='Motor command'))
fig.add_trace(go.Scatter(x=np.linspace(0.2, 0.26, 600), y=np.linspace(0, 1, 600),
                    mode='lines',
                    name='Tangent of the signal'))
fig.add_trace(go.Scatter(x=np.ones(600)*0.26, y=np.linspace(0, 1, 600),
                    mode='lines',
                    name='Reaching Time constant',line=dict(dash='dash',color = 'black')),)
fig.update_layout(
    width=800,  # width in pixels
    height=600,
    xaxis=dict(
        tickmode='array',
        tickvals=[0, 0.2,0.26, 0.4, 0.6, 0.8, 1],  # Specify the tick labels
    )  # height in pixels
)
fig.show()
```
